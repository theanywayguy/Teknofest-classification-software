```python
!pip install roboflow

from roboflow import Roboflow
rf = Roboflow(api_key="")
project = rf.workspace("teknofestdostdusman").project("teknofest-classification")
version = project.version(2)
dataset = version.download("yolov8")
```

```python
from ultralytics import YOLO


model = YOLO("yolov8n.pt")


model.train(
    data="/content/teknofest-classification-2/data.yaml",
    imgsz=640,
    batch=64,
    workers=4,
    cache='disk',
    half=True,         
    amp=True,           
    device='0',         
    pretrained=True,
    epochs=15,
    patience=5,
    optimizer='AdamW',
    lr0=0.001,
    lrf=0.1,
    cos_lr=True,
    name="teknofest-detect",
    plots=True,
    save=True,
    project="runs/detect",
    resume=True
)
```

```python
from ultralytics import YOLO


model = YOLO("/content/best.pt")


results = model.val(
    data="/content/teknofest-classification-2/data.yaml",
    split="test",
    imgsz=640,
    batch=16,
    verbose=True,
    device='0',   
    half=True     
)


print("\nTest set metrics:")
print(f"  Precision: {results.box.p}")
print(f"  Recall:    {results.box.r}")
print(f"  mAP50:     {results.box.map50}")
print(f"  mAP50-95:  {results.box.map}")
```
